```python
# import the excel file
crowdfunding = pd.read_excel("./Starter_Files/Resources/crowdfunding.xlsx")
crowdfunding
```

```python
# get the unique category and create the dataframe
category = crowdfunding['category & sub-category'].str.split('/',expand=True)[0].unique()
categories_df= pd.DataFrame(category,columns=['category'])
categories_df
```

```python
#add the category_id as consecutive
categories_df['category_id']=['cat{}'.format(i+1) for i, _ in enumerate(categories_df['category'])]
categories_df
```

```python
#reorder the columns in the dataframe

categories_df=categories_df[['category_id','category']]
categories_df
```

```python
#export the data frame in a csv

categories_df.to_csv('category.csv',index=False)
```
